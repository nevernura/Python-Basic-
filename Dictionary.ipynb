# A simple dictionary

Consider a game featuring aliens that can have different colors and point
values

```python
alien1 = {'color': 'green', 'points': 5}


#Accessing a dictionary
print(alien1['color']) #Returns value associated with 'color'
print(alien1['points']) #Returns value associated with 'points'
```

### Is the a limit to the number of key-value pairs?

Nope, there is no limited. You can have unlimited numbers of key-value pairs.

### Let's say I shoot alien1 so you'll get some points

```python
print(f"You earned {alien1['points']} points!")
```

### So we can say that once a dictionary have been created, we can pull the value associated with the key (in this case 'points' and 'colors')

# Adding New Key-Value Pairs

Dictionaries are dynamic structures, and you can add new key-value pairs
to a dictionary at any time. To add a new key-value pair, you would give the
name of the dictionary followed by the new key in square brackets, along
with the new value.


Eg, alien1['new_key'] = any value

```python
print(f"Init: {alien1}")

alien1['x_position'] = 0
alien1['y_position'] = 10

print(f"position known: {alien1}")
```

# Modifying Values in a Dictionary

```python
alien1 = {}

alien1 = {'color': 'green'}
print(f"The alien is {alien1['color']} in color.")

#Now I will change the color of the alien
alien1['color'] = 'red'
print(f"Yikes! The alien changed its color to {alien1['color']}.")
```

### Now let's track the movement of the alien

```python
alien1 = {'x_position': 0, 'y_position': 25, 'speed': 'medium'}


print(f"Original position: {alien1['x_position']}")

#Depending on the speed we move the alien to the right
if alien1['speed'] == 'slow':
  x_increment = 1
elif alien1['speed'] == 'medium':
  x_increment = 2
else:
  x_increment = 3

#Calc new position
alien1['x_position'] = alien1['x_position'] + x_increment

print(f"New position: {alien1['x_position']}")
```

# Removing Key-Value Pairs

When you no longer need a piece of information that’s stored in a
dictionary, you can use the del statement to completely remove a key-value
pair. All del needs is the name of the dictionary and the key that you want
to remove.


**Note**: del removes key-value pair permanently

```python
alien1 = {'color': 'green', 'points':5}
print(alien1)

del alien1['color']
print(alien1)
```

# Using get() to Access Values

```python
alien1 = {'color': 'green', 'points':5}


print(alien1['speed'])

```

The get() method requires a key as a first argument. As a second optional
argument, you can pass the value to be returned if the key doesn’t exist:

```python
find_speed = alien1.get('speed', 'No such key found')

print(find_speed)

print(alien1.get('color'))

#If there’s a chance the key you’re asking for might not exist
#use the get() method instead of the square bracket notation.
```

# Looping Through a Dictionary

A single Python dictionary can contain just a few key-value pairs or
millions of pairs. Because a dictionary can contain large amounts of data,
Python lets you loop through a dictionary. Dictionaries can be used to store
information in a variety of ways; therefore, several different ways exist to
loop through them. You can loop through all of a dictionary’s key-value
pairs, through its keys, or through its values.

# Looping Through All Key-Value Pairs

Before we explore the different approaches to looping, let’s consider a new
dictionary designed to store information about a user on a website.

```python
user_0 = {
'username': 'efermi',
'first': 'enrico',
'last': 'fermi',
}

for key, value in user_0.items():
  print(f"\nKey: {key}")
  print(f"Value: {value}")
```

To write a for loop for a dictionary, you create names for the two variables
that will hold the key and value in each key-value pair. You can choose any
names you want for these two variables.

The second half of the for statement includes the name of the dictionary
followed by the method items(), which returns a sequence of key-value
pairs. The for loop then assigns each of these pairs to the two variables
provided. In the preceding example, we use the variables to print each key,
followed by the associated value.

# Looping Through All the Keys in a Dictionary

The keys() method is useful when you don’t need to work with all of the
values in a dictionary.

keys() returns a sequence of all keys

```python
favorite_languages = {
'jen': 'python',
'sarah': 'c',
'edward': 'rust',
'phil': 'python',
}

for name in favorite_languages.keys():
  print(name.title())


print()
#Looping through keys is a default behaviour in dictionary
for name in favorite_languages:
  print(name.title())
```

```python
favorite_languages = {
'jen': 'python',
'sarah': 'c',
'edward': 'rust',
'phil': 'python',
}

print("The following languages have been mentioned:")
for language in favorite_languages.values():
  print(language.title())
```

# Looping Through All Values in a Dictionary

If you are primarily interested in the values that a dictionary contains, you
can use the values() method to return a sequence of values without any
keys.

```python
favorite_languages = {
'jen': 'python',
'sarah': 'c',
'edward': 'rust',
'phil': 'python',
}


print("Languages known are:")
for name in favorite_languages.values():
  print(name.title())
```

# A List in a Dictionary

Say you ordered a pizza with a thick crust and contains the following topping: mushroom and extra cheese

We can represent this pizza using a dictionary and list.

```python
# Store information about a pizza being ordered.
pizza = {
'crust': 'thick',
'toppings': ['mushrooms', 'extra cheese', 'no olives'],
}

# Summarize the order.
print(f"You ordered a {pizza['crust']}-crust pizza with the following toppings:")

for topping in pizza['toppings']:
  print(f"\t{topping}")
```

# A Dictionary in a Dictionary

```python
users = {
      'aeinstein': {
      'first': 'albert',
      'last': 'einstein',
      'location': 'princeton',
                    },
      'mcurie': {
      'first': 'marie',
      'last': 'curie',
      'location': 'paris',
      },
}


for username, user_info in users.items():
  print(f"\nUsername: {username}")
  full_name = f"{user_info['first']} {user_info['last']}"
  location = user_info['location']
  print(f"\tFull name: {full_name.title()}")
  print(f"\tLocation: {location.title()}")
```

Q. Consider two lists:

name = ['alic', 'bob', 'cat', 'dick', 'eerie', 'flan']

marks = [60, 30, 50, 70, 20, 90]

Create a dictionary where the keys are names and items are marks

### Some more dictionary examples

```python
name = ['alic', 'bob', 'cat', 'dick', 'eerie', 'flan']
marks = [60, 30, 50, 70, 20, 90]

report = dict(zip(name, marks))
#zip keyword combine elements from multiple iterables
#iterables must be of same lengths

print(report)
```

Q. Using the above data, classify pass or fail with pass being >= 40.

```python
result = {name: 'Pass' if marks >= 40 else 'Fail' for name, marks in report.items()}

print(result)

res = {name: 'Pass' if marks > 40 else 'Fail' for index, (name, marks) in enumerate(report.items())}
print(res)
```
