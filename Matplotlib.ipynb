### In this tutorial, we will work with the Iris dataset

```python
import seaborn as sns

# Load the famous 'iris' flower dataset
df = sns.load_dataset('iris')

# View the first 5 rows
print(df.head())
```

### We will use the sepal width and length as our data for plotting

# Basic Plot

```python
import matplotlib.pyplot as plt

x = df['sepal_length']
y = df['sepal_width']

plt.plot(x, y) # Creates a canvas where x and y is plotted
plt.xlabel("Sepal Length (cm)") # Labels the x axes
plt.ylabel("Sepal Width (cm)") # Label the y axes
plt.show() # This line displays the output to the console
```

Let's now create 6 subplots (2 rows and 3 cols)

```python
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(12,8))

ax[0][0].plot(x, x)
ax[0][0].set_title("Linear")
ax[0][1].plot(x, x**2)
ax[0][1].set_title("Square")
ax[0][2].plot(x, x**3)
ax[0][2].set_title("Cube")
ax[1][0].plot(x, np.sin(x))
ax[1][0].set_title("Sine")
ax[1][1].plot(x, np.cos(x))
ax[1][1].set_title("Cosine")
ax[1][2].plot(x, np.exp(x))
ax[1][2].set_title("Exponential")
#ax.grid(True) - This would generate an error

#This works for making grids in all
# ax variable is a 2-dimensional NumPy array
# To iterate through it in an easier way, use ax.flat
for a in ax.flat:
    a.grid(True)

fig.suptitle("6 subplots", fontsize=16)
plt.show()
```

# Bar Plots

The primary arguments for plt.bar() are:

1. x: A sequence of scalars representing the x-coordinates or categories of the bars.

2. height: A scalar or sequence of scalars representing the height(s) of the bars.

3. width: The width(s) of the bars (default is 0.8).

4. bottom: The y-coordinate(s) of the bars' bases (default is 0).

5. align: Alignment of the bars to the x-coordinates ('center' or 'edge').

```python
"""
For this, we will find the petal length of each species
A bar chart with a categorical x-axis and
continuous y-axis requires aggregation
(e.g., mean, count, sum) for the y-values within each category.
"""



new_df = df.groupby('species')['petal_length'].mean().reset_index()


new_df.head()
```

```python
x = new_df['species']
y = new_df['petal_length']


plt.bar(x, y)
plt.xlabel("Iris Species", fontsize=16)
plt.ylabel("Average Petal Length (cm)", fontsize=16)
plt.show()
```

```python
plt.barh(x, y)
plt.xlabel("Iris Species", fontsize=16)
plt.ylabel("Average Petal Length (cm)", fontsize=16)
plt.show()
```

```python
c = ["yellow", "red", "green"]

plt.bar(x, y, color=c, width=0.6, edgecolor='black')
plt.xlabel("Iris Species", fontsize=16)
plt.ylabel("Average Petal Length (cm)", fontsize=16)
plt.show()
```

# One Dimension Random Walk

X[t] = X[t-1] + Z[t]

t = 1, 2, 3, . . . n

X[0] = 0, Z[t] = N(0, 1)

```python
z = np.random.normal(0, 1, 1200)
x = np.zeros(1200)
for i in range(1, 1200):
    x[i] = x[i-1] + z[i]

#One random walk
plt.plot(x, linewidth=.5, color='r')
plt.show()
```

### 200 random walk paths

```python
# z needs shape (200, 1200) for 200 simulations
z = np.random.normal(0, 1, (200, 1200)) #200 rows, 1200 points
x = np.zeros((200, 1200))

# Random walk calculation
for j in range(200):
  for i in range(1, 1200):
      x[j][i] = x[j][i-1] + z[j][i]

# Plot each row (each walk)
for j in range(200):
    plt.plot(x[j], linewidth=.5, color='r', alpha=0.12)

plt.title("200 Random Walks")
plt.show()
```
