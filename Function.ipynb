# Recursive Functions

```python
#Factorial of a number using loops

def fact_loop(n):

  fact = 1
  while(n > 0):
    fact = fact * n
    n = n - 1

  return fact

print(fact_loop(5))
```

5 x 4 x 3 x 2 x 1 1! == 1, 0! == 1

```python
#Factorial using recursion

def fact_rec(n):

  if n == 1:
    return 1

  return n * fact_rec(n-1)

print(fact_rec(5))
```

```python
#String reversal
#Loops

def reverse_string(s):
  reverse = ""
  for char in s:
    reverse = char + reverse
  return reverse


print(reverse_string("hello"))
```

```python
#String reversal
#Recursion

def reverse_string(s):

    #Base case
    if len(s) == 1:
      return s

    return s[-1] + reverse_string(s[:-1])

reverse_string("hello")

```

## We can also create Switch Case using Lambda functions

```python
import math

def get_square(x):
    return x * x

def get_cube(x):
    return x * x * x

def get_sqrt(x):
    return math.sqrt(x)


operations = {
    "square": get_square,
    "cube": get_cube,
    "sqrt": get_sqrt
}


num = 16
result = operations["sqrt"](num)
print(result)
```

```python
operations = {
    "square": lambda x: x ** 2,
    "cube": lambda x: x ** 3,
    "sqrt": lambda x: x ** (5/7)
}


num = 16
result = operations["sqrt"](num)
print(result)
```

## Use case 1 - map()

```python
#Print the squares from 1 - 10

def square(x):
  return x * x

nums = [i for i in range(1,11)]

for num in nums:
  print(square(num), end=" ")
```

```python
#Using map and lambda function

#map - Applies a function on each element of an iterable
#This case, lambda function is applied on nums

nums = [i for i in range(1,11)]

squares = list(map(lambda x: x**2, nums))

print(squares)
```

## Use case 2 - filter

filter - extracts elements from an iterable (such as a list, tuple, or string) that satisfy a given condition

```python
def is_even(x):
  if x % 2 == 0:
    return x

nums = [i for i in range(1,11)]
print(nums)

for num in nums:
  if is_even(num) != None:
    print(num, end=" ")
```

```python
#Using lambda functions

nums = [i for i in range(1,11)]
print(nums)

evens = list(filter(lambda x: x%2 == 0, nums))

print(evens)
```

# Generator Function

```python
def simple_generator():
    print("Step 1: Preparing data...")
    yield "Data Piece A"

    print("Step 2: Processing more...")
    yield "Data Piece B"

    print("Step 3: Finishing up...")
    yield "Data Piece C"

# Create the generator object
my_gen = simple_generator()

# Use 'next' to trigger each yield
print(next(my_gen))
# print(next(my_gen))
# print(next(my_gen))
```

```python
def countdown(n):
    while n > 0:
        yield n
        n -= 1

timer = countdown(3)
print(next(timer))
print("Waiting...")
print(next(timer))
```
