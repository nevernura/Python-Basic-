Find factorial of a number

5! = 5 * 4 * 3 * 2 * 1

```python
n = int(input("Enter number: "))

i = n
fact = i

while(i >= 1):
    fact = fact * i
    i -= 1

print(f"{n}! = {fact}")
```

### Print multiples of a number but skip the 5th multiple

**Use case of continue**

```python
num = int(input("Enter number: "))

for i in range(1, 11):
    if i == 5:
      continue
    print(f"{num} * {i} == {num*i}")
```

### Can we reverse a string without slicing it

```python
s = input("")

for i in range(len(s)-1, -1, -1):
  print(s[i], end="")
```
