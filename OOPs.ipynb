# Demonstrating instance variable

In Python, we usually define these inside a special method called __init__ (the "constructor") using the keyword self.

```python
class Car:
    def __init__(self, color, model):
        # These are instance variables
        self.color = color
        self.model = model

# Creating two separate 'instances'
car1 = Car("Red", "Sedan")
car2 = Car("Blue", "SUV")

# Changing car1 does NOT change car2
car1.color = "Green"

print(car1.color) # Output: Green
print(car2.color) # Output: Blue (remains unchanged)
```

# Constructor

constructor is a special method that is automatically called when you create a new instance of a class. Its primary job is to initialize the instance variables we just talked about.

In Python, the constructor is always named __init__.

Notice how we don't have to manually call __init__; it happens just by writing Car().

```python
class Car:
    # This is the Constructor
    def __init__(self, brand, year):
        print(f"Creating a new {brand}...")
        self.brand = brand  # Setting initial state
        self.year = year    # Setting initial state

# The moment we run this line, the constructor fires off!
my_car = Car("Tesla", 2024)

print(my_car.brand) # Output: Tesla
print(my_car.year)
```

# What does self do?

In OOP, self is how an object refers to itself. It’s the way an object accesses its own unique data (instance variables) without getting confused by other objects of the same class. Rather, self sort of assigns an identity to an instance/object.


3 Rules of self:
It’s the first parameter: You must include it in the definition of any method inside a class (like __init__(self, ...)).

It links data to the instance: Using self.variable_name ensures the data stays with that specific object.

Python handles the "passing": You don't pass self when you call the method; Python does that part for you automatically

# Is it always required to use self?

The short answer is: No, but it depends on what the method needs to do.

If a method needs to touch the "soul" of a specific object—like checking a user's balance or changing a car's color—it must use self.

However, sometimes you want a function to live inside a class simply because it's related to the topic, even if it doesn't need to know anything about a specific instance.

# Encapsulation

```python
class BankAccount:
  def __init__(self, balance):
    self.__balance = balance

  def deposit(self, amount):
    if amount > 0:
      self.__balance += amount
      print(f"Deposited {amount}. New balance: {self.__balance}")
    else:
      print("Invalid deposit amount!")

  def get_balance(self):
    return self.__balance


account = BankAccount(100)

#We cannot access this private variable __balance
#account.__balance = 50000

#But we can access deposit
account.deposit(50)
print(account.get_balance)
```

# len() function in the background

```python
class CustomContainer:
    def __init__(self, items):
        self.items = items

    def __len__(self):
        """Returns the number of items in the container."""
        return len(self.items) # Calls the built-in len() on the internal list

my_container = CustomContainer([1, 2, 3, 4, 5])
print(len(my_container)) # Output: 5
```

```python
class Squarearea:
  def __init__(self, length, breadth):
    self.length = length
    self.breadth = breadth

  def area(self):
    return self.length * self.breadth

 #This helps me handle more than two squares
  def __add__(self, other):
    new_area = self.area() + other.area()
    return Squarearea(new_area, 1)

  def __str__(self):
    return f"Area of Square = {self.area()}"

#Now try to add the area of more than squares
s1 = Squarearea(12,3)
s2 = Squarearea(1,2)
s4 = Squarearea(1,2)

s3 = s1 + s2 + s4
print(s3)
```

# Polymorphism via Inheritance (Method Overriding)

Even though all objects are treated as Animal, the actual method executed is from the actual
class (Dog, Cat, etc.).

This is classic class-based polymorphism.

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        # We define a generic behavior or leave it empty
        raise NotImplementedError("Subclass must implement abstract method")

class Dog(Animal):
    def speak(self):
        return f"{self.name} says: Woof! Woof!"

class Cat(Animal):
    def speak(self):
        return f"{self.name} says: Meow..."

# --- The Polymorphic Power ---

def animal_concert(animals):
    """A single function that handles any animal type."""
    for animal in animals:
        # This is the 'Many Forms' in action.
        # Python looks up the correct 'speak' at runtime.
        print(animal.speak())

# Creating our objects
pets = [Dog("Buddy"), Cat("Whiskers"), Dog("Rex")]

animal_concert(pets)
```

# Using Polymorphic Functions with Different Classes



```python
class Duck:
    def make_sound(self):
        return "Quack! Quack!"

class Person:
    def make_sound(self):
        return "Hello, my name is John."

# The Polymorphic Function
def sound_test(anything_that_can_make_sound):
    """This function is 'blind' to class types.
    It only cares if the object has a make_sound method."""
    print(anything_that_can_make_sound.make_sound())

# Creating unrelated objects
daisy = Duck()
john = Person()

# Both work!
sound_test(daisy)
sound_test(john)
```
